```python
import statsmodels.api as sm
from statsmodels.tsa.api import VAR
from statsmodels.tsa.stattools import adfuller
import datetime
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./../data/df.csv')
df = df.drop('Unnamed: 0', axis = 1)
df['btc'] = df['btc']/100
df
```

```python
from statsmodels.tsa.base.datetools import dates_from_str

quarterly = df["year"].astype(str) + "Q" + df["quarter"].astype(str)
quarterly = dates_from_str(quarterly)
btc = df[['btc','testimony','minutes','beige','speech1','speech2']]
btc.index = pd.DatetimeIndex(quarterly)
btc
```

```python
btc2 = btc.query("index >= '2014'")
btc2
```

```python
## VAR 모델 시작

model = VAR(btc)
results = model.fit(1) ## 시차 수 2개 - 올해, 작년, 제작년 총 3기간 설정
results.summary()
```

```python
results.plot() # 시계열 데이터 시각화
```

```python
results.plot_acorr() # autocorrelation 그래프 : 점선을 넘지 않아야 좋음
```

```python
# 최적 lag수 구하기

model.select_order(10)
results = model.fit(maxlags=10, ic='aic')
results, results.k_ar
```

```python
lag_order = 2
results.forecast(btc.values[-lag_order:], 5) 
```

```python
results.plot_forecast(4) # forecast 시각화 - 향후 10개 분기의 값들을 예측하겠다.
```

```python
# 각 변수가 1이 상승하는 충격을 받았을때 각 변수들의 예상 반응 경로
irf = results.irf(10) # 10기간 : 10개 분기
irf.plot(orth=False)
```

```python
## VAR 모델 시작

model = VAR(btc2)
results = model.fit(1) ## 시차 수 2개 - 올해, 작년, 제작년 총 3기간 설정
results.summary()
```

```python
results.plot() # 시계열 데이터 시각화
```

```python
lag_order = 1
results.forecast(btc.values[-lag_order:], 5)
results.plot_forecast(4) # forecast 시각화 - 향후 10개 분기의 값들을 예측하겠다.
```
